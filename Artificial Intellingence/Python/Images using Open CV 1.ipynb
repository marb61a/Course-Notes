```python
# OpenCV will need to be installed, it can be installed using
# pip install opencv-python
import cv2
import matplotlib.pyplot as plt
```

```python
# Read an image in using cv2, this will read in an array of rgb values
# Each image will be a matrix that is 3 layers deep
img = cv2.imread('beach.jpg')
```

```python
# This shows the number of columns, rows and depth
print(img.shape)
```

```python
# Using Matplotlib to render an image
# OpenCV interprets images in BGR format
plt.imshow(img)
```

```python
# Creating a seperate image to map color channels properly
rgb_image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
# This will map properties correctly
# With the axis turned off
plt.axis('off')
plt.imshow(rgb_image)
```

```python
# Cropping a part of the image
# Positions are also switched when using cropping
# The second position corresponds to rows, the first to 
# y, the third value takes in the channel, if all use :
cropped_image = rgb_image[1000:2000,1000:2000,:]
plt.imshow(cropped_image)
```

```python
# Resize an image
newCropImage = cv2.resize(cropped_image, (100, 100))
```

```python
# Show the new cropped image
# Will be of less quality
plt.imshow(newCropImage)
```
